```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading data

df1 = pd.read_csv(r'C:\Users\aksha\OneDrive\Desktop\movies.csv')
```

```python
# Viewing data

df1.head()
```

```python
# Checking for missing data 

for col in df1.columns:
    pct_missing=np.mean(df1[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# Checking data types of columns

df1.dtypes
```

```python
# Creating new column to capture 'release' year

df1['Released_Year'] = df1['released'].astype(str).str[7:12]
```

```python
df1.head()
```

```python
df1 = df1.drop('Released_Year', axis=1)
```

```python
df1.head()
```

```python
df1['Released_year'] = df1['released'].astype(str).str.split().str[2]
```

```python
df1.head()
```

```python
# looking at highest grossing films 

# df1.sort_values(by=['gross'], inplace =False, ascending= False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Converting gross to numeric: 

df1['gross'] = pd.to_numeric(df1['gross'], errors ='coerce').fillna(0).astype(int)
```

```python
df1.head()
```

```python
# df1.sort_values(by=['gross'], inplace =False, ascending= False)
```

```python
# Dropping missing & negative values 
df1 = df1.dropna()
df1 = df1 [df1['gross'] >=0]
```

```python
#Checking for correlations
#Chances are that budget & company have a high correlation with gross revenue
#Checking this out

plt.scatter(x=df1['budget'], y =df1['gross'])

plt.title('Budget vs Gross revenue')

plt.xlabel('Budget')

plt.ylabel('Gross revenue')

plt.show()
```

```python
#Plotting regression plot using seaborn to visually see correlation

sns.regplot(x='budget' , y ='gross', data = df1, scatter_kws={"color" : "Blue"}, line_kws={"color" : "red"})
```

```python
df1.corr(method = 'pearson') # there are other methods like Kendall, Spearman
```

```python
correlation_matrix = df1.corr(method = 'pearson')
```

```python
sns.heatmap(correlation_matrix, annot = True)
```

```python
df_numerized = df1


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr().head()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
highly_corr = sorted_pairs[(sorted_pairs)> 0.5]
```
